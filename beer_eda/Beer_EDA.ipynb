# Interactive Table

```python
brand_list = th['Brand Name']
@interact(Brand= brand_list)
def filter_brand(Brand):
    return th[th['Brand Name']== Brand]
```

# Interactive Chart

```python
brand_list2 = brand_year['Brand Name']
@interact(brand1=brand_list2, brand2=brand_list2)
def plot_chart(brand1, brand2):
    df=brand_year[brand_year['Brand Name'].isin([brand1, brand2])]
    sns.barplot(x='Market Share',y='Brand Name',hue='Year', data=df)
    plt.grid()
```
